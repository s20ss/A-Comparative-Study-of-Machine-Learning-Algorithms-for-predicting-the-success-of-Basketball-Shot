```python
colss=['FINAL_MARGIN','SHOT_NUMBER','PERIOD','SHOT_CLOCK','PTS','CLOSE_DEF_DIST','SHOT_DIST','TOUCH_TIME','DRIBBLES','SHOT_RESULT']
```
